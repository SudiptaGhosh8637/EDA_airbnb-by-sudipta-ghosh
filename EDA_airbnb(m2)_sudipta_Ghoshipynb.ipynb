### Variables Description

a) id: Unique serial number.

b) name: Description given to each accomodation.

e) neighbourhood_group: Various boroughs(town/district) within New York city.

h) room_type: Variety of rooms depending on the size.

i) price: Cost of the rooms.

j) minimum_nights: Number of nights, hosts stay in that accomodation.

k) number_of_reviews: Number of times hosts give reviews.

l) last_review: Date of last review.

m) reviews_per_month: Ratio of number of reviews to number of days in each month.

n) calculated_host_listings_count: Number of times a host visited that particular room.

## 3. ***Data Wrangling***

```python
# Write your code to make your dataset analysis ready.

# Drop duplicate rows, if any

airbnb_df.drop_duplicates(inplace= True)
```

```python
# Check for null values in each column

airbnb_df.isna().sum()
```

We can remove null values of host_name as we can't know the authenticity of the data and also it won't affect much of the dataframe. Also we can replace null values of reviews_per_month with 0.

```python
# Remove null values from host_name
airbnb_df.dropna(axis=0, subset=['host_name'], inplace=True)
```

```python
# Replacing null values of reviews_per_month with 0

airbnb_df.reviews_per_month.fillna(0, inplace= True)
```

```python
# Removing columns which doesn't contribute much in exploration of data and storing in new dataframe

new_airbnb_df = airbnb_df.drop(['id', 'host_id', 'name', 'last_review'], axis = 1)
```

```python
# Checking for null values after cleaning , if any

new_airbnb_df.isna().sum()
```

```python
# Checking for infinite valuees, if any

new_airbnb_df[new_airbnb_df == np.inf].sum()
```

```python
#How dataset look after cleaning
plt.figure(figsize=[6,4])
sns.heatmap(new_airbnb_df.isnull())
plt.show()
```

The above graph shows that there is no null values in the dataset

```python
#check the column with numeric data types
numeric_var=new_airbnb_df.select_dtypes(include=['int64','float64']).columns
numeric_var
```

```python
#show the distribution of numeric column of both old and new dataset
plt.figure(figsize=[20,6])
for i,var in enumerate(numeric_var):
  plt.subplot(2,4,i+1)
  sns.histplot(airbnb_df[var],bins=20,kde=True)
  sns.histplot(new_airbnb_df[var],bins=20,kde=True)
```

# **Another ascept of cleaning data is to remove outliers.**

```python
# Variables needs to be checked for outliers

features_to_be_analysed = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
```

```python
# Checking the variables with outliers
plt.figure(figsize=[15,5])
for index, col in enumerate(features_to_be_analysed):
  plt.subplot(2,3,index+1)
  plt.boxplot(new_airbnb_df[col],showmeans=True,sym='r')
  plt.ylabel(col)
```

The data which is outside the 1.5*IQR, (where, IQR = inter-quartile range (25 - 75) percentile) are considered as an outlier. In this dataset, we can't say any particular value as outlier cause it can be genuine depending on the customer behaviour. So, we do not remove any data value which is seen as outlier.

Outlier detection is only useful when we need a decision boundary to distinguish between various categories but in EDA we don't need to do this.

# d. Price

Lets hover over the price column and see its distribution and nature

```python
# show the distribution of the price

plt.figure(figsize=(5,6))
sns.histplot(new_airbnb_df['price'],color="y")
plt.show()
```

As we see. the price distribution has a positively skewed tail. We can get an intuition from the above graph that prices of most rooms is between 0 and $1000. Let's try to normalize the inputs using log function, so that it can viewed more clearly.

```python
plt.figure(figsize=(10,5))
sns.histplot(np.log(new_airbnb_df[~(new_airbnb_df['price']==0)]['price']), color="skyblue",kde=True)
plt.show()
```

Now, lets make groups of prices as follows..

price 0 to 100 = 0

price 100 to 200 = 1

price > 200 = 2

```python
# Number of prices in each price group

new_airbnb_df['price_group'] = new_airbnb_df['price'].apply(lambda x : 0 if x<100 else (1 if 100<= x < 200 else 2))
new_airbnb_df['price_group'].value_counts()
```

```python
# Percentage of room alloted to people with price less than 200

print(f'Percentage of room alloted to people with price less than 200: {round((21867+17224)/(21867+17224+9783)*100,2)}%')
```

Following code can be interpreted as, there are..

21867 rooms having price less than 100

17224 rooms havind price between 100 to 200

9783 rooms having price greater than 200

Most rooms are below 200 price range with almost 80% and very few are more than 200.

```python
# Representation of price groups

df1 = pd.DataFrame(new_airbnb_df.groupby(['price_group'])['price_group'].value_counts().sort_values(ascending = False))
mylabels = ["<100", "100 to 200", ">200"]
df1.plot.pie(subplots=True,labels = mylabels, shadow= True, colors= ['yellowgreen', 'lightblue', 'gold'], explode= [0.0, 0.1, 0.1], figsize=(8,6))
plt.legend(title = "Price:")
plt.title('Price group in pie chart')

plt.show()
```

Let's check how many nights spent depending on the prices of the room

```python
# Distribution of prices with nights spents

plt.figure(figsize=[12, 5])
plt.scatter(new_airbnb_df['price'], new_airbnb_df['minimum_nights'],  color = 'y')
plt.ylabel('nights spent',fontsize=15)
plt.xlabel('prices',fontsize=15)
plt.show()
```

This looks like an interesting plot.

Most nights spent are for rooms having prices below 200. Very few nights are spent for room with price over 2000. It is also quite logical that people living in highly expensive accomodation might not stay for good number of days.

Manhattan and Brooklyn are two most visited destinations as we have found out earlier. We are going to find which neighbourhood in these neighbourhood_group is most expensive.

```python
# What are the costliest neighbourhood in Manhattan

n_df = pd.DataFrame(new_airbnb_df.groupby(['neighbourhood_group', 'neighbourhood'])['price'].mean().Manhattan.nlargest(10))
ax = n_df.plot.bar(ylabel = 'Average price', title = 'Top 10 neighbourhood in Manhattan according to mean cost', figsize = (6,4), color = 'y')
```

```python
# What are the least costliest neighbourhood in Manhattan

n1_df = pd.DataFrame(new_airbnb_df.groupby(['neighbourhood_group', 'neighbourhood'])['price'].mean().Manhattan.nsmallest(5))
ax = n1_df.plot.bar(ylabel = 'Average price', title = 'bottom 5 neighbourhood in Manhattan according to mean cost', figsize = (6,4), color = 'r')
```

Tribeca is the most costliest neighbourhood in Manhattan with mean price of 490.6, whereas other neighbourhoods have almost similar mean price.

Inwood, Marble Hill and Washington Heights being least expensive neighbourhood in Manhattan.

```python
# What are the costliest neighbourhood in Brooklyn

n2_df = pd.DataFrame(new_airbnb_df.groupby(['neighbourhood_group', 'neighbourhood'])['price'].mean().Brooklyn.nlargest(10))
ax = n2_df.plot.bar(ylabel = 'Average price', title = 'Top 10 neighbourhood in Brooklyn according to mean cost', figsize = (6 , 4), color = 'Lime')
```

Sea Gate is the most costliest neighbourhood in Brooklyn with mean price of 487.9, whereas other neighbourhoods have almost similar mean price.

There are various conclusion we can draw from this feature:

1- Almost 80% people likely to stay at rooms with price below 200.

2- Duration of stay in less expensive rooms are more compared to more expensive rooms.

3- There is more likely of people choosing less expensive neighbourhood for their accomodation even in expensive neighbourhood_groups like Manhattan and Brooklyn.
